```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def char2png(char):
    plt.figure(figsize=(4, 4))
    plt.text(0.5, 0.5, char[0], fontsize=400, ha='center', va='center')
    plt.axis('off')

    file_path='matplot_offset.png'
    plt.savefig(file_path, dpi=1000, format='png', bbox_inches='tight', pad_inches=0.1)
    # plt.show()

    return file_path
```

```python
char = 'Q'
char2png(char)
!start ./matplot_offset.png
```
